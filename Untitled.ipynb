```python
print("Before", train_df.shape, test_df.shape, combine[0].shape, combine[1].shape)

train_df = train_df.drop(["Ticket", "Cabin"], axis=1)
test_df = test_df.drop(["Ticket", "Cabin"], axis=1)
combine = [train_df, test_df]

print("After", train_df.shape, test_df.shape, combine[0].shape, combine[1].shape)
```

```python
for dataset in combine:
    dataset["Title"] = dataset.Name.str.extract(' ([A-Za-z]+)\.', expand=False) ## Title은 직함, 정규표현식 잘 모르겠음

pd.crosstab(train_df['Title'], train_df['Sex'])
```

```python
for dataset in combine:
    dataset["Title"] = dataset["Title"].replace(["Lady", "Countess", "Capt", "Col", 'Don', "Dr", "Major", "Rev",\
                                                "Sir", "Jonkheer", "Dona"], "Rare")
    dataset["Title"] = dataset["Title"].replace("Mlle", "Miss")
    dataset["Title"] = dataset["Title"].replace("Ms", "Miss")
    dataset["Title"] = dataset["Title"].replace("Mme", "Mrs")

train_df[["Title", "Survived"]].groupby(["Title"], as_index=False).mean()
```

```python
title_mapping = {"Mr": 1, "Miss": 2, "Mrs": 3, "Master": 4, "Rare": 5}
for dataset in combine:
    dataset["Title"] = dataset["Title"].map(title_mapping) ## df.map(dict) 이용하면 값 쉽게 변환할 수 있는듯
    dataset["Title"] = dataset["Title"].fillna(0) ## na 값 0으로 채우기
    
train_df.head()
```

```python
train_df = train_df.drop(["Name", "PassengerId"], axis=1)
test_df = test_df.drop(["Name"], axis=1) # 얘는 나중에 아이디로 해답이랑 비교할라고 그러나
combine = [train_df, test_df]
train_df.shape, test_df.shape
```

```python
for dataset in combine:
    dataset["Sex"] = dataset["Sex"].map( {"female": 1, "male": 0}).astype(int) ## map()으로 값변환, astype()으로 type변환

train_df.head()
```

```python
# grid = sns.FacetGrid(train_df, col="Pclass", hue="Gender")
grid = sns.FacetGrid(train_df, row="Pclass", col="Sex", size=2.2, aspect=1.6)
grid.map(plt.hist, "Age", alpha=.5, bins=20)
grid.add_legend()
```

```python
guess_ages = np.zeros((2, 3))
guess_ages
```

```python
for dataset in combine:
    for i in range(0, 2): ## median값 구하려고 빈칸 일단 없애는 중
        for j in range (0, 3):
            guess_df = dataset[(dataset["Sex"] == i) & \
                              (dataset["Pclass"] == j+1)]["Age"].dropna()
            
            # age_mean = guess_df.mean()
            # age_std = guess_df.std()
            # age_guess = rnd.uniform(age_mean - age_std, age_mean + age_std)
            
            age_guess = guess_df.median()
            
            # Convert random age float to nearest .5 age
            guess_ages[i, j] = int( age_guess/0.5 + 0.5 ) * 0.5 ## 1.3 => 1.5, 1.5 => 2

    for i in range(0, 2): ## 빈칸에 나이 집어넣음
        for j in range(0, 3):
            dataset.loc[ (dataset.Age.isnull()) & (dataset.Sex == i) & \
                       (dataset.Pclass == j+1), "Age"] = guess_ages[i,j]
        
    dataset["Age"] = dataset["Age"].astype(int)

train_df.head()
```

```python
train_df["AgeBand"] = pd.cut(train_df["Age"], 5) ## cut이라는 method 이용하면 구간별로 나눌 수 있음, 5부분으로 나눈다
train_df[["AgeBand", "Survived"]].groupby(["AgeBand"], as_index=False).mean().sort_values(by="AgeBand", ascending=True)
```

```python
for dataset in combine: # Age 행 값을 모두 범위에 해당하는 카테고리값(0 ~ 3) 으로 변환
    dataset.loc[ dataset["Age"] <= 16, "Age"] = 0
    dataset.loc[(dataset["Age"] > 16) & (dataset["Age"] <= 32), "Age"] = 1
    dataset.loc[(dataset["Age"] > 32) & (dataset["Age"] <= 48), "Age"] = 2
    dataset.loc[(dataset["Age"] > 48) & (dataset["Age"] <= 64), "Age"] = 3
    dataset.loc[ dataset["Age"] > 64, "Age"]
train_df.head()
```

```python
## notebook에서 not이 아니라 now 인듯(We can not remove the AgeBand feature)

train_df = train_df.drop(["AgeBand"], axis=1)
combine = [train_df, test_df]
train_df.head()
```

```python
for dataset in combine:
    dataset["FamilySize"] = dataset["SibSp"] + dataset["Parch"] + 1 ## 그냥 이렇게 더할 수 있음, 자신포함: +1
    
train_df[["FamilySize", "Survived"]].groupby(["FamilySize"], as_index=False)\
.mean().sort_values(by="Survived", ascending=False)
```

```python
for dataset in combine: ## IsAlone 열 만듦
    dataset["IsAlone"] = 0
    dataset.loc[dataset["FamilySize"] == 1, "IsAlone"] = 1

train_df[["IsAlone", "Survived"]].groupby(["IsAlone"], as_index=False).mean()
```

```python
train_df = train_df.drop(["Parch", "SibSp", "FamilySize"], axis=1) ## IsAlone 빼고 가족관계 자료 다 없앰
test_df = test_df.drop(["Parch", "SibSp", "FamilySize"], axis=1)
combine = [train_df, test_df]

train_df.head()
```

```python
for dataset in combine: # 나이하고 좌석등급하고 합침
    dataset["Age*Class"] = dataset.Age * dataset.Pclass

train_df.loc[:, ["Age*Class", "Age", "Pclass"]].head(10) ##loc첫번째 인수?는 열이름, 두번째 인수는 행이름(들)
```

```python
freq_port = train_df.Embarked.dropna().mode()[0] ## mode()가 뭔지 모르겠음, 가장 많이 나온거 추려내는 듯? 0 대신 1넣으면 에러뜸
freq_port
```

```python
for dataset in combine: ## 빈 값 2개는 최빈값인 S로 채운다음 승선지별 생존률을 구함
    dataset["Embarked"] = dataset["Embarked"].fillna(freq_port)

train_df[["Embarked", "Survived"]].groupby(["Embarked"], as_index=False)\
.mean().sort_values(by="Survived", ascending=False)
```

```python
for dataset in combine: ## 승선지 값을 숫자 카테고리로 바꾸고 타입도 object에서 메모리 적게먹는 int로 바꿈
    dataset["Embarked"] = dataset["Embarked"].map( {"S": 0, "C": 1, "Q": 2}).astype(int)
    
train_df.head()
```
